# Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import bokeh.palettes as palette
```

```python
col = palette.viridis(7)
columns = ["N","T_N"]
```

## Well-mixed population

### Linear fitness case

```python
df1 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau0.0625.txt', header=None, names=columns)
df2 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau0.125.txt' , header=None, names=columns)
df3 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau0.25.txt'  , header=None, names=columns)
df4 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau0.5.txt'   , header=None, names=columns)
df5 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau1.0.txt'   , header=None, names=columns)
df6 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau2.0.txt'   , header=None, names=columns)
df7 = pd.read_csv('./Sim_A/Fig2/wm_lfc_tau4.0.txt'   , header=None, names=columns)

def scaling_rule(N, tau, sigma):
    return 1 / (2 * tau * sigma**2) * (np.log(N)**2)

n = np.logspace(1.8,8,num=int(1e3))
teor_4 = scaling_rule(n,4.0,0.25)
teor_2 = scaling_rule(n,2.0,0.25)

plt.plot(df1['N'], df1['T_N'], '-o', linewidth=0.5, markersize=4, color = col[0], label=r'$\tau=1/16$')
plt.plot(df2['N'], df2['T_N'], '-o', linewidth=0.5, markersize=4, color = col[1], label=r'$\tau=1/8$')
plt.plot(df3['N'], df3['T_N'], '-o', linewidth=0.5, markersize=4, color = col[2], label=r'$\tau=1/4$')
plt.plot(df4['N'], df4['T_N'], '-o', linewidth=0.5, markersize=4, color = col[3], label=r'$\tau=1/2$')
plt.plot(df5['N'], df5['T_N'], '-o', linewidth=0.5, markersize=4, color = col[4], label=r'$\tau=1$')
plt.plot(df6['N'], df6['T_N'], '-o', linewidth=0.5, markersize=4, color = col[5], label=r'$\tau=2$')
plt.plot(df7['N'], df7['T_N'], '-o', linewidth=0.5, markersize=4, color = col[6], label=r'$\tau=4$')

plt.plot(n, teor_4, '--', alpha=0.7, color='gray') #, label=r'$\frac{1}{2 \tau \sigma^2} (\log N)^2$')
plt.plot(n, teor_2, '--', alpha=0.7, color='gray')
plt.xlabel(r'$N$')
plt.ylabel(r'$T_N$')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--", alpha=0.4)
plt.legend(loc='upper right')

plt.savefig("Fig2a.pdf", bbox_inches='tight')
plt.show()
```

```python
y1 = df1['T_N']/df1['N']
y2 = df2['T_N']/df2['N']
y3 = df3['T_N']/df3['N']
y4 = df4['T_N']/df4['N']
y5 = df5['T_N']/df5['N']
y6 = df6['T_N']/df6['N']
y7 = df7['T_N']/df7['N']

tau_list = [1.0/16.0 , 1.0/8.0, 1.0/4.0, 1.0/2.0, 1.0, 2.0, 4.0]
sigma = 0.25

x1 = tau_list[0]*(sigma**2)*df1['N']
x2 = tau_list[1]*(sigma**2)*df2['N']
x3 = tau_list[2]*(sigma**2)*df3['N']
x4 = tau_list[3]*(sigma**2)*df4['N']
x5 = tau_list[4]*(sigma**2)*df5['N']
x6 = tau_list[5]*(sigma**2)*df6['N']
x7 = tau_list[6]*(sigma**2)*df7['N']

plt.plot(x1, y1, '.', color = col[0], label=r'$\tau=1/16$')
plt.plot(x2, y2, '.', color = col[1], label=r'$\tau=1/8$')
plt.plot(x3, y3, '.', color = col[2], label=r'$\tau=1/4$')
plt.plot(x4, y4, '.', color = col[3], label=r'$\tau=1/2$')
plt.plot(x5, y5, '.', color = col[4], label=r'$\tau=1$')
plt.plot(x6, y6, '.', color = col[5], label=r'$\tau=2$')
plt.plot(x7, y7, '.', color = col[6], label=r'$\tau=4$')

plt.xlabel(r'$\tau \sigma^2 N$')
plt.ylabel(r'$T_N/N$')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--", alpha=0.4)
plt.legend()

plt.savefig("Fig2b.pdf", bbox_inches='tight')
plt.show()
```
